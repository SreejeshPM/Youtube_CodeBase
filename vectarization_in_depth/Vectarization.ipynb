### 3. Word2Vec - Embeddings


#### A: Skip-Gram

“The quick brown fox jumps over the lazy dog”

<ul>
<li>As it can be seen that input is a 10,000-dimensional vector given our vocabulary size =10,000, containing a 1 corresponding to the position of our target word. 
<li>The output layer consists of 10,000 neurons with the Softmax activation function applied, so as to obtain the respective probabilities against every word in our vocabulary.
<li>Now the most important part of this network, the hidden layer is a linear layer i.e. there’s no activation function applied there, and the optimized weights of this layer will become the learned word embeddings.
<li>For example, let’s say we decide to learn word embeddings with the above network. In that case, the hidden layer weight matrix shape will be M x N, where M = vocabulary size (10,000 in our case) and N = hidden layer neurons (300 in our case).
<li>Once the model gets trained, the final word embedding for our target word will be given by the following calculation:<br>
 <b>1×10000 input vector * 10000×300 matrix = 1×300 vector</b>

300 hidden layer neurons were used by Google in their trained model, however, this is a hyperparameter and can be tuned accordingly to obtain the best results.
</ul>

#### B. CBOW

<ul>
<li>The dimension of our hidden layer and output layer stays the same as the skip-gram model.
<li>However, just as we read that a CBOW model takes context words as input, here the input is C context words in the form of a one-hot encoded vector of size 1xV each, where V = size of vocabulary, making the entire input CxV dimensional.
<li>Now, each of these C vectors will be multiplied with the Weights of our hidden layer which are of the shape VxN, where V = vocab size and N = Number of neurons in the hidden layer.
<li>If you can imagine, this will result in C, 1xN vectors, and all of these C vectors will be averaged element-wise to obtain our final activation for the hidden layer, which then will be fed into our output softmax layer.
<li>The learned weight between the hidden and output layer makes up the word embedding representation.
<ul>

<h4> When to use skip-gram model and CBOW? </h4> 

The skip-gram works well with <b>small datasets and can better represent rare words.<br>
CBOW is found to train faster than skip-gram and can better represent frequent words.</b> <br>
So the choice of skip-gram VS. CBOW depends on the kind of problem that we’re trying to solve.

```python
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt

def plot_vector(model):
    vocab = list(model.wv.key_to_index)
    X = model.wv[vocab]
    tsne = TSNE(n_components=2)
    X_tsne = tsne.fit_transform(X)
    df = pd.DataFrame(X_tsne, index=vocab, columns=['x', 'y'])
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.scatter(df['x'], df['y'])
    for word, pos in df.iterrows():
        ax.annotate(word, pos)
```

### 4. GloVe
Global Vectors for Word Representation

<ul>
    <li>Word2Vec is a window-based method, in which the model relies on local information for generating word embeddings, which in turn is limited to the <b>adjudged window size</b> that we choose.
<li>This means that the semantics learned for a target word is only affected by its surrounding words in the original sentence, which is a somewhat inefficient use of statistics, as there’s a lot more information we can work with.
<li>GloVe on the other hand captures both <B>global and local statistics</B> in order to come up with the word embeddings.

<h4> When to use GloVe </h4>
<ul>
<li>GloVe has been found to outperform other models on <b>word analogy, word similarity, and Named Entity Recognition tasks,</b> so if the nature of the problem you’re trying to solve is similar to any of these, GloVe would be a smart choice.
<li>Since it incorporates global statistics, it can capture the semantics of rare words and performs well even <b>on a small corpus.</b>
</ul>

```python
import numpy as np
from scipy import spatial

embeddings_dict={}
with open('glove_6B/glove.6B.50d.txt','rb') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict[word] = vector
        
def find_closest_embeddings(embedding):
    return sorted(embeddings_dict.keys(), key=lambda word:
spatial.distance.euclidean(embeddings_dict[word], embedding))
```

```python
embeddings_dict[b'warming']
```

```python
find_closest_embeddings(embeddings_dict[b'warming'])[:10]
```
